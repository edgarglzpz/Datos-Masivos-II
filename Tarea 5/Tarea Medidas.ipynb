```python
import pandas as pd
```

```python
catalogo = pd.read_excel("catalogo_municipios_sepomex.xlsx", engine = "openpyxl")
```

```python
catalogo["Llave"] = catalogo["DESC_ESTADO"]+catalogo["DESC_MUNICIPIO"]
```

```python
sucursales_x_municipio = pd.read_excel("sucursales_por_municipio.xlsx", engine = "openpyxl")
```

```python
sucursales_x_municipio
```

```python
num_mun = len(set(catalogo.Llave))
print("Número de municipios en el país:", num_mun)
```

```python
num_mun_suc = len(set(sucursales_x_municipio.DESC_MUN))
print("Número de municipios con sucursales en el país:", num_mun_suc)
```

```python
#Los 3 bancos con mayor número de sucursales
num_sucursales = sucursales_x_municipio.groupby("DESC_INSTITUCION")["NUM_SUCURSALES"].sum().sort_values(ascending=False).to_frame("Número de sucursales")
print("Los 3 bancos con mayor número de sucursales:")
num_sucursales.head(3)
```

```python
print("Los 3 bancos con menor número de sucursales:")
num_sucursales.tail(16)
```

```python
#Los 3 bancos con presencia en mayor número de municipios
num_sucursales_presencia = sucursales_x_municipio.groupby("DESC_INSTITUCION")["DESC_INSTITUCION"].count().sort_values(ascending=False).to_frame("Presencia en municipios")
print("Los 3 bancos con presencia en mayor número de municipios:")
num_sucursales_presencia.head(3)
```

```python
#¿Qué banco es el que ayuda a que haya mayor inclusión financiera en el país?
mun_una_suc = sucursales_x_municipio[sucursales_x_municipio["NUM_SUCURSALES"] == 1]
```

```python
mun_una_suc.groupby("DESC_INSTITUCION")["DESC_INSTITUCION"].count().sort_values(ascending=False).to_frame("Presencia única en municipios")
```
